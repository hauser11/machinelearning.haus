```python
from tensorflow.keras.datasets import boston_housing
(x_train, y_train), (x_test,y_test) = boston_housing.load_data()
```

```python
from tensorflow.keras import models,layers
nn=models.Sequential()
nn.add(layers.Dense(8,activation='relu',input_dim=13))
```

```python
nn.add(layers.Dense(1,activation='relu'))
```

```python
nn.compile(optimizer='rmsprop',loss='mse',metrics=['mae','mse'])
```

```python
history=nn.fit(x_train, y_train, epochs=100)
```

```python
from matplotlib import pyplot
pyplot.plot(history.history['mse'])
pyplot.ylabel("MSE")
pyplot.xlabel("Epoch")
pyplot.show()
```

```python
nn.evaluate(x_test,y_test)
```

```python
y_pred = nn.predict(x_test)
y_pred
```

```python
pc = models.Sequential()
pc.add(layers.Dense(1,activation='relu',input_dim=13))
pc.compile(optimizer='rmsprop',loss='mse',metrics=['mae','mse'])
history=pc.fit(x_train, y_train, epochs=200)
pc.evaluate(x_test,y_test)
```

```python
nn2= models.Sequential()
nn2.add(layers.Dense(8,activation='relu'))
nn2.add(layers.Dense(1,activation='relu'))
nn2.compile(optimizer='rmsprop', loss='mse',metrics=['mae','mse'])
history=nn2.fit(x_train, y_train, epochs=300)
nn2.evaluate(x_test,y_test)
```
