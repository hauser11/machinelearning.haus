```python
import tensorflow
import numpy
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from keras.utils import to_categorical
from tensorflow.keras import models, layers
import pandas

def main():
    data = pandas.read_csv("/Users/MattHauser/Desktop/splice.csv")
    x = data.values[:,:61]
    y = data.values[:,61:62]
    ohe = OneHotEncoder(handle_unknown='ignore')
    ohe.fit(x)
    x = ohe.transform(x).toarray()
    
    oe= OrdinalEncoder()
    oe.fit(y)
    y=oe.transform(y)
    y=to_categorical(y)
    
    fold_auc=[]
    folds=10
    for fold in range(folds):
        x_train = numpy.array([])
        y_train = numpy.array([])
        x_test = numpy.array([])
        y_test = numpy.array([])
        
        print("Fold =",fold)
        
        print("Partitioning the data into train and test.")
        
        for i in range(len(x)):
            if i % folds == fold:
                x_test = numpy.append(x_test,x[i])
                y_test = numpy.append(y_test,y[i])
                
            else:
                x_train = numpy.append(x_train,x[i])
                y_train = numpy.append(y_train,y[i])
                
        x_train = x_train.reshape(len(x_train) // len(x[0]), len(x[0]))
        y_train = y_train.reshape(len(y_train) // len(y[0]), len(y[0]))
        x_test = x_test.reshape(len(x_test) // len(x[0]), len(x[0]))
        y_test = y_test.reshape(len(y_test) // len(y[0]), len(y[0]))

        print("Training")
        model = models.Sequential()
        model.add(layers.Dense(61, activation='relu', input_dim=len(x[0])))
        model.add(layers.Dense(3, activation='softmax'))
        model.compile(optimizer='rmsprop', loss = 'categorical_crossentropy',
                     metrics=[tensorflow.keras.metrics.AUC()])
        history=model.fit(x_train, y_train, epochs=100, verbose=0)

        print("Testing")
        s= model.evaluate(x_test, y_test)
        fold_auc.append(s[1])
        print("AUC =", s[1])
        print("Average AUC =", s[1])

        
    print("average auc:", numpy.mean(fold_auc))
main()
        
```
